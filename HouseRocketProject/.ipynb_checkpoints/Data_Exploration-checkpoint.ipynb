```python
# Leitura e concatenação dos dados de treino e teste em um conjunto único
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
price = pd.read_csv('./data/sample_submission.csv')
test['SalePrice'] = price.SalePrice
houses = pd.concat([train, test])
houses.drop('Id', axis=1, inplace=True)
```

## Pré-Processamento dos Dados

```python
data_book = pd.read_csv('./data/HousePricesDataBook.csv')
```

```python
colunas_remover = list(data_book[data_book.Expectativa == 'Baixo']['Variável'])
```

```python
houses.drop(colunas_remover, axis=1, inplace=True)
```

```python
houses.drop_duplicates()
```

```python
df_faltantes = utils.verifica_faltantes(houses)
df_faltantes
```

```python
houses = utils.remove_faltantes(houses, columns=['Electrical', 'GarageCars', 'GarageArea'], to_remove='linhas')
df_faltantes = utils.verifica_faltantes(houses)
houses = utils.remove_faltantes(houses, df_faltantes.index)
```

```python
utils.verifica_faltantes(houses).shape[0]
```

```python
houses['HouseAge'] = houses['YrSold'] - houses['YearBuilt']
```

```python
houses['CentralAir'].replace({'N': 0, 'Y': 1}, inplace=True)
```

```python
utils.corr_to_target(houses, 'SalePrice')
```

```python
houses.drop(['KitchenAbvGr', 'OverallCond', 'YrSold'], axis=1, inplace=True)
```
